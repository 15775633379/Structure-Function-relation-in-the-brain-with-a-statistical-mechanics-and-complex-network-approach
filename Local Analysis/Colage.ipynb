```python
from PIL import Image, ImageDraw
from os.path import expanduser, join
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
home_figuras = '/gdrive/Shareddrives/GENSyR /CNF (Complejidad y NeuroFiﾌ《ica)/Carpeta de trabajo/Tesis Juan y Diego /Figuras/analisis local/'
home_resultados = '/gdrive/Shareddrives/GENSyR /CNF (Complejidad y NeuroFiﾌ《ica)/Carpeta de trabajo/Tesis Juan y Diego /Figuras/finales/'

sujetos = ['Emotion_F+','Emotion_F-','Emotion_M+','Emotion_M-','Social_F+','Social_F-','Social_M+','Social_M-','WM_F+','WM_F-','WM_M+','WM_M-']
matrices = ['_estructural_','_funcional_']
tareas = ['EMOTION','SOCIAL','WM']
graficas = ['nodos_1.png','nodos_2.png']
```

```python
iss = [0,500]
jss = [0,1000]
for tarea in tareas:
  new = Image.new('RGBA', (2000,1000))
  for i in range(2):
    for j in range(2):
      ruta = home_figuras + tarea + matrices[i] + graficas[j]
      img = Image.open(ruta)

      img = img.resize((1000,500))

      new.paste(img,(jss[j],iss[i]))
  new.save(home_resultados + tarea + '_nodos_local_PE.png')
```

```python
jss = [0,900]
for tarea in tareas:
  new = Image.new('RGBA', (2200,1800))
  for i in range(2):
    for j in range(2):
      ruta = home_figuras + tarea + matrices[i] + graficas[j]
      img = Image.open(ruta)

      img = img.resize((2200,900))

      new.paste(img,(0,jss[j]))
    new.save(home_resultados + tarea + '_' + matrices[i] + '_nodos_local_PE.png')
```

```python
iss = [0,1000]

for sujeto in sujetos:
  new = Image.new('RGBA', (2000,500))
  for i in range(2):
    ruta = home_figuras + sujeto + graficas[i]
    img = Image.open(ruta)
    img = img.resize((1000,500))

    new.paste(img,(iss[i],0))
  new.save(home_resultados + sujeto + '.png')
```
