```python
df = pd.read_csv('/gdrive/Shareddrives/GENSyR /CNF (Complejidad y NeuroFísica)/Carpeta de trabajo/Datos Proyecto Conectoma Humano (HCP)/Datos_Sujetos.csv')
df["Subject"] = df["Subject"].astype(str)
df = df[np.invert(df["Subject"].isin(excepciones + ['192136','195950','116524', '581349']))]
df = df[df["3T_Full_MR_Compl"] == True]
```

```python
# Tareas por las que se quieren seleccionar los datos
filtros = ['WM', 'Emotion', 'Social']
# Elementos por categoria que se quieren analizar
categoria = 'Gender'
sexos = ['M','F']

# Caracteristicas de los sujetos que se quieren extraer
#features = ['Subject', 'Gender', 'Age', 'WM_Task_Acc', 'Dexterity_Unadj']
#features = ['Subject', 'Gender', 'Age','Endurance_AgeAdj','WM_Task_Acc','WM_Task_Median_RT','Language_Task_Acc','Language_Task_Median_RT','Emotion_Task_Face_Acc','Emotion_Task_Face_Median_RT','CogFluidComp_AgeAdj','ProcSpeed_AgeAdj', 'NEOFAC_A','NEOFAC_O','NEOFAC_C','NEOFAC_N','NEOFAC_E','Social_Task_TOM_Perc_TOM']
features_personalidad = ['NEOFAC_A','NEOFAC_O','NEOFAC_C','NEOFAC_N','NEOFAC_E']
features_Adj = ['Endurance_AgeAdj','CogFluidComp_AgeAdj','ProcSpeed_AgeAdj']
features_RT = ['WM_Task_Median_RT', 'Emotion_Task_Median_RT', 'Social_Task_TOM_Median_RT_TOM'] 
features_Acc = ['WM_Task_Acc', 'Emotion_Task_Acc', 'Social_Task_TOM_Perc_TOM']

features = ['Subject', 'Gender', 'Age'] + features_personalidad + features_Adj + features_RT + features_Acc
```

# Se extraen las listas de los sujetos de las poblaciones

```python
# Se extrae un subdiccionario con las caracteristicas requeridas
data = df[features]
# Se eliminan del diccionario ya filtrado con las features las variables NaN que puedan existir
data = data.dropna(axis=0)
# Se excluyen para el analisis las personas con edades mayores a 36 años
data = data[data["Age"] != "36+"]

for index, filtro in enumerate(filtros):

  data[filtro] = data[features_Acc[index]]/data[features_RT[index]]
  data = data.drop(columns=[features_Acc[index], features_RT[index]])
```

```python
data
```

```python
sujetos = {}


for filtro in filtros:
  for sexo in sexos:

    # Se filtran los datos por la categoria
    data_aux = data[data['Gender'] == sexo]
    data_aux = data_aux.sort_values(by=filtro, ascending=False)

    subjects_ordenados = data_aux["Subject"]
    lista = list(subjects_ordenados)

    # Se extraen y guardan en terminos de la tarea filtro y la categoria los 25 nodos que mas tiempo de reaccion tienen
    name = filtro + '_' + sexo + '+' 
    sujetos[name] = lista[0]

    # Se extraen y guardan en terminos de la tarea filtro y la categoria los 25 nodos que menor tiempo de reaccion tienen
    name = filtro + '_' + sexo + '-' 
    sujetos[name] = lista[len(lista)-1]


name = 'Sujetos.npy'
fname = join(fdatapoblaciones, 'Sujetos Tesis', name)
np.save(fname, sujetos)

sujetos
```

# Se extraen los datos psicometricos de cada sujeto

```python
# Se extrae un subdiccionario con las caracteristicas requeridas
data = df[features]
# Se eliminan del diccionario ya filtrado con las features las variables NaN que puedan existir
data = data.dropna(axis=0)
# Se excluyen para el analisis las personas con edades mayores a 36 años
data = data[data["Age"] != "36+"]

for index, filtro in enumerate(filtros):

  data[filtro] = data[features_Acc[index]]/data[features_RT[index]]
  data = data.drop(columns=[features_Acc[index], features_RT[index]])

data = data.set_axis(list(data['Subject']), axis = 'index')
data = data.drop(columns=['Subject'])
data = data.drop(columns=['Gender'])
data = data.drop(columns=['Age'])

data = data.div(data.max(axis='index'), axis='columns')

name = 'Sujetos.npy'
fname = join(fdatapoblaciones, 'Sujetos Tesis', name)
arr_dic = np.load(fname, allow_pickle=True)
dic = arr_dic.item()

data_sujetos = data.loc[list(dic.values())]
data_sujetos = data_sujetos.set_axis(list(dic.keys()), axis = 'index')
data_sujetos = data_sujetos.rename_axis("Subject")

name = 'Sujetos_psicometricos.xlsx'
fname = join(fdatapoblaciones, 'Sujetos Tesis', name)
data_sujetos.to_excel(fname, index = list(dic.keys()))
```
